```python
# 🚀 Task 2: Exploratory Data Analysis (EDA) - Titanic Dataset

# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# Step 2: Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Preview data
df.head()
```

```python
# Step 3: Basic Info
print("Shape of dataset:", df.shape)
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 4: Summary Statistics
df.describe(include="all")
```

```python
# Step 5: Histograms for numeric features
df.hist(figsize=(12, 8), bins=20, edgecolor="black")
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Step 6: Boxplots (to detect outliers)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[["Age", "Fare"]])
plt.title("Boxplots of Age and Fare")
plt.show()
```

```python
# ✅ Step 7: Correlation Heatmap (numeric only)
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=[np.number])  # keep only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 8: Pairplot for selected features
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived")
plt.show()
```

```python
# Step 9: Insights / Observations
print("🔎 Insights:")
print("1. Many passengers have missing Age values.")
print("2. Fare distribution is highly skewed with some extreme outliers.")
print("3. Survival seems correlated with Pclass (1st class more likely to survive).")
print("4. Females had a higher survival rate compared to males.")
print("5. Younger passengers tended to survive more compared to older ones.")
```
